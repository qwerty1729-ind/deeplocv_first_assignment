```python
import cv2
import matplotlib.pyplot as plt

def capture_image():
    cap = cv2.VideoCapture(0)  # 0 is the default webcam

    if not cap.isOpened():
        print("Error: Could not access the camera.")
        return None

    # Capture a single frame
    ret, frame = cap.read()
    if ret:
        cv2.imwrite("captured_image.jpg", frame)  # Save the image to file
        print("Image saved as 'captured_image.jpg'")

        # Display the captured image
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))  # Convert to RGB for correct color
        plt.title("Captured Image")
        plt.axis('off')  # Hide axes
        plt.show()
    else:
        print("Error: Could not capture an image.")

    # Cleanup
    cap.release()
    cv2.destroyAllWindows()

# Run the function
capture_image()
```

```python
import cv2
import matplotlib.pyplot as plt
def grey_scaling(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load as grayscale
    plt.imshow(image, cmap='gray')  # Show grayscale image
    plt.title("Grayscale Image")
    plt.axis('off')  # Hide axes
    plt.show()

# Call the function
grey_scaling("captured_image.jpg")
```

```python
import cv2
import matplotlib.pyplot as plt

def threshold_binary(image_path, threshold_value):
    # Load the grayscale image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply binary thresholding
    _, binary_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

    # Display the thresholded image
    plt.imshow(binary_image, cmap='gray')
    plt.title("Thresholding")
    plt.axis('off')
    plt.show()

# Call the function
threshold_binary("captured_image.jpg", 70)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def threshold_16_gray_levels(image_path):
    # Load the grayscale image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Divide the grayscale range into 16 levels
    gray_16_levels = (image // 16) * 16  # Quantize the values into 16 levels

    # Display the reduced grayscale image
    plt.imshow(gray_16_levels, cmap='gray')
    plt.title("16 Grey Levels")
    plt.axis('off')
    plt.show()

# Call the function
threshold_16_gray_levels("captured_image.jpg")
```

```python
import cv2
import matplotlib.pyplot as plt
def sobel_and_canny_edge_detection(image_path):
    # Load the grayscale image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    # Apply Sobel filter in x and y directions
    sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Sobel in X
    sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Sobel in Y
    sobel_combined = cv2.magnitude(sobel_x, sobel_y)       # Magnitude of gradients
    # Apply Canny edge detector
    canny_edges = cv2.Canny(image, threshold1=100, threshold2=200)
    # Display results
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(image, cmap='gray')
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(sobel_combined, cmap='gray')
    plt.title("Sobel Edge Detection")
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(canny_edges, cmap='gray')
    plt.title("Canny Edge Detection")
    plt.axis('off')

    plt.show()

# Call the function
sobel_and_canny_edge_detection("captured_image.jpg")
#observed that canny edge is better
```

```python
import cv2
import matplotlib.pyplot as plt

def apply_gaussian_filter(image_path):
    # Load the grayscale image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply Gaussian blur with a 5x5 kernel
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

    # Display the blurred image
    plt.imshow(blurred_image, cmap='gray')
    plt.title("Gaussian Blur")
    plt.axis('off')
    plt.show()

# Call the function
apply_gaussian_filter("captured_image.jpg")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def sharpen_image(image_path):
    # Load the grayscale image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply Gaussian blur first
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

    # Define a sharpening kernel
    sharpening_kernel = np.array([[0, -1, 0],
                                  [-1, 5,-1],
                                  [0, -1, 0]])

    # Apply the sharpening filter
    sharpened_image = cv2.filter2D(blurred_image, -1, sharpening_kernel)

    # Display the sharpened image
    plt.imshow(sharpened_image, cmap='gray')
    plt.title("Sharpened Image")
    plt.axis('off')
    plt.show()

# Call the function
sharpen_image("captured_image.jpg")
```

```python
import cv2
import matplotlib.pyplot as plt

def convert_rgb_to_bgr(image_path):
    # Load the image in RGB using Matplotlib (default is RGB)
    image_rgb = plt.imread(image_path)

    # Convert RGB to BGR using OpenCV
    image_bgr = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR)

    # Display the converted image using Matplotlib (convert it back to RGB for display)
    plt.imshow(cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB))  # Convert back to RGB for Matplotlib
    plt.title("Converted Image (RGB to BGR)")
    plt.axis('off')
    plt.show()

# Call the function
convert_rgb_to_bgr("captured_image.jpg")
```
