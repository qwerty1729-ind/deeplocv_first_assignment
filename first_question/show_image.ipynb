```python
import matplotlib.pyplot as plt

def show(name, n, m, i, Title):
    plt.subplot(n, m, i)
    plt.imshow(name, cmap='gray')
    plt.title(Title)
    plt.axis('off')
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image (replace 'sample_image.jpg' with your image name)
image = cv2.imread('sample_image.jpg', cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread('sample_image2.jpg', cv2.IMREAD_GRAYSCALE)
image3 = cv2.imread('sample_image3.jpg', cv2.IMREAD_GRAYSCALE)

# Display the image
show(image, 1,3, 1, "batman")
show(image2, 1,3, 2, "ironman")
show(image3, 1,3, 3, "Mr.bean")
# Call the function
plt.show()
```
