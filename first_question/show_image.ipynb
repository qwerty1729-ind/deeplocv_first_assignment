```python
import matplotlib.pyplot as plt

def show(name, n, m, i, Title):
    plt.subplot(n, m, i)
    plt.imshow(name, cmap='gray')
    plt.title(Title)
    plt.axis('off')
```

```python
import cv2
import matplotlib.pyplot as plt

# Load images in grayscale
image1 = cv2.imread('sample_image.jpg', cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread('sample_image2.jpg', cv2.IMREAD_GRAYSCALE)
image3 = cv2.imread('sample_image3.jpg', cv2.IMREAD_GRAYSCALE)

# Invert the grayscale images
image1 = 255 - image1
image2 = 255 - image2
image3 = 255 - image3

# Call the show function (without changing it)
show(image1, 1, 3, 1, "batman")
show(image2, 1, 3, 2, "ironman")
show(image3, 1, 3, 3, "Mr.bean")

plt.show()
```
